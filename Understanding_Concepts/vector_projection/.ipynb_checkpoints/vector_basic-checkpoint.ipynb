```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
np.random.seed(678)

plt.figure(figsize=(12, 12))
n_samples = 2000
random_state = 192
X, y = make_blobs(n_samples=n_samples, random_state=random_state,centers =2)
plt.figure(figsize=(12, 12))
plt.scatter(X[:,0],X[:,1],c=y)
plt.grid()
plt.title("Generated Fake Data")
plt.show()
```

```python
# Lets Create a simple Vector (coordinate -2.5, 5)
classification = np.asarray([1,2.5])
X_added = np.vstack((X,classification))
y_added = np.insert(y,y.shape[0],3)
plt.figure(figsize=(12, 12))
plt.scatter(X_added[:,0],X_added[:,1],c=y_added)
plt.grid()
plt.title("Generated Fake Data with Vector")
plt.show()
```

```python
mag = np.sqrt(classification.dot(classification))
print(classification.dot([0.5,3.5])/mag)
print(classification.dot([0.5,0.5])/mag)

predicted_c = []
for current_x in X:
    value = classification.dot(current_x)/mag
    if value < 2.5:
        predicted_c.append(0)
    else:
        predicted_c.append(1)

predicted_c = np.asarray(predicted_c)
predicted_c = np.insert(predicted_c,predicted_c.shape[0],3)
plt.figure(figsize=(12, 12))
plt.scatter(X_added[:,0],X_added[:,1],c=predicted_c)
plt.grid()
plt.title("Generated Fake Data with Vector")
plt.show()
```

```python
x = np.linspace(-5, 8, 100)
y = np.linspace(-5, 12.5, 100)
xv, yv = np.meshgrid(x, y)
xv = xv.flatten()
yv = yv.flatten()

color_change = []
for xx,yy in zip(xv,yv):
    value = classification.dot([xx,yy])/mag
    if value < 2.5:
        color_change.append('red')
    else:
        color_change.append('blue')
xv = np.insert(xv,xv.shape[0],1)
yv = np.insert(yv,yv.shape[0],2.5)
xv = np.insert(xv,xv.shape[0],0)
yv = np.insert(yv,yv.shape[0],0)
color_change.append('yellow')
color_change.append('black')

plt.figure(figsize=(12, 12))
plt.scatter(xv,yv,c=color_change)
plt.show()
```

```python
import pandas as pd
import matplotlib
import seaborn as sns

df = pd.read_csv('../../Dataset/kaggle_Pima_Indians_Diabetes/Pima_Indians_Diabetes.csv')
df.head()
```

```python
# Draw the heatmap with the mask and correct aspect ratio
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
sns.heatmap(df.corr(), vmax=.3, center=0,square=True, annot=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
df.columns
```

```python
df_features = df[['Glucose','BMI']]
df_outcome = df[['Outcome']]
print(df_features.head())
print(df_features.info())
```

```python
print(df_outcome.head())
print(df_outcome.info())
```

```python
df_features_np = df_features.values
df_outcome_np = df_outcome.values
plt.figure(figsize=(12, 12))
plt.grid()
plt.scatter(df_features_np[:,0],df_features_np[:,1],c=np.squeeze(df_outcome.values))
plt.show()
```

```python
classification = np.asarray([125,35])
df_features_np = np.vstack((df_features_np,classification))
df_outcome_np = np.insert(df_outcome_np,df_outcome_np.shape[0],3)
plt.figure(figsize=(12, 12))
plt.scatter(df_features_np[:,0],df_features_np[:,1],c=df_outcome_np)
plt.grid()
plt.show()
```

```python
mag = np.sqrt(classification.dot(classification))

predicted_c = []
for current_x in df_features.values:
    value = classification.dot(current_x)/mag
    if value < 130:
        predicted_c.append(0)
    else:
        predicted_c.append(1)

predicted_c = np.asarray(predicted_c)
predicted_c = np.insert(predicted_c,predicted_c.shape[0],3)

print(predicted_c.shape)
print(df_features_np.shape)

plt.figure(figsize=(12, 12))
plt.scatter(df_features_np[:,0],df_features_np[:,1],c=predicted_c)
plt.grid()
plt.title("Generated Fake Data with Vector")
plt.show()
```
