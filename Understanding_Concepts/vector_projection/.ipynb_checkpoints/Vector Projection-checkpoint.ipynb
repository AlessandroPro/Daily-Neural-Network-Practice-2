```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import interactive
interactive(True)

sns.set(style="white")
np.random.seed(5678)
```

```python
df = pd.read_csv('../../Dataset/kaggle_Pima_Indians_Diabetes/Pima_Indians_Diabetes.csv')
```

```python
df.head()
```

```python
# Draw the heatmap with the mask and correct aspect ratio
fig = matplotlib.pyplot.gcf()
fig.set_size_inches(18.5, 10.5)
sns.heatmap(df.corr(), vmax=.3, center=0,square=True, annot=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```python
df.columns
```

```python
df_features = df[['Glucose','BMI','Age']]
df_outcome = df[['Outcome']]
print(df_features.head())
print(df_features.info())
```

```python
print(df_outcome.head())
print(df_outcome.info())
```

```python
%matplotlib notebook
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
colors = {1:'red', 0:'blue'}
ax.scatter(df_features['Glucose'], df_features['BMI'], c=df_outcome['Outcome'].apply(lambda x: colors[x]))
plt.show()
```
