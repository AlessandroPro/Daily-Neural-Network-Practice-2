```python
import numpy as np
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
np.random.seed(678)

plt.figure(figsize=(12, 12))
n_samples = 2000
random_state = 192
X, y = make_blobs(n_samples=n_samples, random_state=random_state,centers =2)
plt.figure(figsize=(12, 12))
plt.scatter(X[:,0],X[:,1],c=y)
plt.grid()
plt.title("Generated Fake Data")
plt.show()
```

```python
# Lets Create a simple Vector (coordinate -2.5, 5)
classification = np.asarray([1,2.5])
X_added = np.vstack((X,classification))
y_added = np.insert(y,y.shape[0],3)
plt.figure(figsize=(12, 12))
plt.scatter(X_added[:,0],X_added[:,1],c=y_added)
plt.grid()
plt.title("Generated Fake Data with Vector")
plt.show()
```

```python
mag = np.sqrt(classification.dot(classification))
print(classification.dot([0.5,3.5])/mag)
print(classification.dot([0.5,0.5])/mag)

predicted_c = []
for current_x in X:
    value = classification.dot(current_x)/mag
    if value < 2.5:
        predicted_c.append(0)
    else:
        predicted_c.append(1)

predicted_c = np.asarray(predicted_c)
predicted_c = np.insert(predicted_c,predicted_c.shape[0],3)
plt.figure(figsize=(12, 12))
plt.scatter(X_added[:,0],X_added[:,1],c=predicted_c)
plt.grid()
plt.title("Generated Fake Data with Vector")
plt.show()
```

```python
x = np.linspace(-5, 8, 100)
y = np.linspace(-5, 12.5, 100)
xv, yv = np.meshgrid(x, y)
xv = xv.flatten()
yv = yv.flatten()

print(xv.shape)
print(yv.shape)

color_change = []
for xx,yy in zip(xv,yv):
    value = classification.dot([xx,yy])/mag
    if value < 0:
        color_change.append(1)
    else:
        color_change.append(0)
        
plt.figure(figsize=(12, 12))
plt.scatter(xv,yv,c=color_change)
plt.show()
```
