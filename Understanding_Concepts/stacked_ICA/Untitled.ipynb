```python
import logging
from time import time

from numpy.random import RandomState
import matplotlib.pyplot as plt

from sklearn.datasets import fetch_olivetti_faces
from sklearn.cluster import MiniBatchKMeans
from sklearn.decomposition import FastICA
```

```python
dataset = fetch_olivetti_faces(shuffle=True, random_state=2)
faces = dataset.data
```

```python
print(faces.shape)
```

```python
iica = FastICA(n_components=100)
```

```python
small=iica.fit_transform(faces.T).T
print(small.shape)
```

```python
fig=plt.figure(figsize=(10, 10))
columns = 10
rows = 10
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(small[i-1].reshape(64,64),cmap='gray')
plt.show()
```

```python
one = FastICA(n_components=200)
```

```python
one_result = one.fit_transform(faces.T).T
print(one_result.shape)
```

```python
two = FastICA(n_components=50)
```

```python
two_results = two.fit_transform(one_result[:50,:].T).T
print(two_results.shape)
```

```python
fig=plt.figure(figsize=(10, 10))
columns = 10
rows = 10
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(one_result[i-1].reshape(64,64),cmap='gray')
plt.show()
```

```python
fig=plt.figure(figsize=(10, 10))
columns = 10
rows = 10
for i in range(1, columns*rows +1):
    fig.add_subplot(rows, columns, i)
    plt.imshow(two_results[i-1].reshape(64,64),cmap='gray')
plt.show()
```
