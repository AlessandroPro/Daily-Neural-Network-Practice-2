```python
import pandas as pd
import numpy as np
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from matplotlib import pyplot as plt
import seaborn as sns
import plotly.graph_objs as go
init_notebook_mode(connected=True)
%matplotlib inline
```

```python
df = pd.read_csv("HospitalGeneralInformation.csv",encoding="ISO-8859-1")
df.describe()
```

```python
df.info()
print('------- How many Null Item per col-------')
df.isnull().sum()
```

```python
df_by_city = df.groupby(['State']).size()
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = df_by_city.index,
        z = df_by_city.astype(float),
        locationmode = 'USA-states',
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = "Number of Hospitals")
        ) ]

layout = dict(
        title = '4,000 Medicare-certified hospitals across the country (US)',
        geo = dict(
            scope='USA',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
iplot(fig)
```

```python
data = [
    go.Bar(
        x=df_by_city.index,
        y=df_by_city
    )
]

layout = go.Layout(
    title='4,000 Medicare-certified hospitals across the country Per State',
    xaxis=dict(
        title='State Name',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Number of hospitals',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```
